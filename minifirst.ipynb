```python
while True:
    total=0
    for i in range(3):
         user=float(input("enter the item price:"))
         total=total+user
    print("total:",+total)
    if user == 0:
        break
    if total>=1000:
        discount=total*0.10
    elif total>=500:
        discount=total*0.05
    else:
        discount=0
    final=total-discount
    print("discount:",+discount)
    print("after removing discount:",+final)
    user=input("do you want to continue?(yes/no):")
    if user == "no":
        break
```
