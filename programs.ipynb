```python
#square pattern
for i in range(4):
    for j in range(4):
        print("*",end=" ")
    print()
```

```python
#right angle triangle
for i in range(4):
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
#inverted triangle
for i in range(4,0,-1):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
#number triangle
for i in range(1,5):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#same number row pattern
for i in range(1,5,1):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
#pyramid pattern
for i in range(1,5):
    print(" "*(5-i), end="")  
    for j in range(i):
        print("* ",end="")      
    print()
```

```python
#reverse number
for i in range(4,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```
