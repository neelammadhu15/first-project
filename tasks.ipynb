```python
#smart employee id genertor
name=input("EnterEmployeeName:")
year=input("EnterJoiningYear:")
department=input("EnterDepartmentName:")
name_part=name[:3].upper()
year_part=year[-2:]
dept_part=department[:3].upper()
employee_id=name_part+year_part+dept_part
print("GeneratedEmployeeID:",employee_id)
```

```python
#smart password validator
password=input("EnterPassword:")
print("CheckingPassword:",password)
upper=False
lower=False
digit=False
special=False
digit_count=0
for ch in password:
    if ch.isupper():
        upper=True
    elif ch.islower():
        lower=True
    elif ch.isdigit():
        digit=True
        digit_count+=1
    else:
        special=True
length=len(password)
i=0
length_ok=False
while i<length:
    i+=1
if length>=8:
    length_ok=True
print("First3chars:",password[:3])
print("Last3chars:",password[-3:])
if len(password)>2:
    masked=password[0]+"*"*(len(password)-2)+password[-1]
else:
    masked=password
print("MaskedPassword:",masked)
print("TotalDigits:",digit_count)
print("UppercaseLetters:",end="")
for ch in password:
    if ch.isupper():
        print(ch,end="")
print()
half=len(password)//2
print("ReverseFirstHalf:",password[:half][::-1]+password[half:])
if password[0].isupper() and password[-1].isdigit():
    print("StartsWithUppercaseAndEndsWithDigit:Yes")
else:
    print("StartsWithUppercaseAndEndsWithDigit:No")
if upper and lower and digit and special and length_ok:
    print("Status:StrongPassword")
else:
    print("Status:WeakPassword")
```

```python
#right aliggned increasing triangle
rows=5
for i in range(1,rows+1):
    for j in range(rows-i):
        print(" ",end="")
    for k in range(i):
        print("*",end="")
    print()
```

```python
#right aligned decreasing triangle
rows=5
for i in range(rows,0,-1):
    for j in range(rows-i):
        print(" ",end="")
    for k in range(i):
        print("*",end="")
    print()
```

```python
#atm withdrawl system
balance=float(input("EnterBalance:"))
amount=float(input("EnterWithdrawalAmount:"))
try:
    if amount<=0:
        raise ValueError("InvalidAmount")
    if amount>balance:
        raise ValueError("InsufficientBalance")
    balance-=amount
    print("RemainingBalance:",balance)
except ValueError as e:
    print(e)
finally:
    print("ThankyouforusingourATM.")
```

```python
#student result processing system
class FailException(Exception):
    pass
marks=int(input("EnterMarks:"))
try:
    if marks<0 or marks>100:
        raise ValueError
    if marks<35:
        raise FailException
except ValueError:
    print("InvalidMarks")
except FailException:
    print("StudentFailed")
else:
    print("StudentPassed")
finally:
    print("ResultProcessingCompleted")
```
